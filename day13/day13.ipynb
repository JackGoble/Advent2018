<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Each time a cart has the option to turn (by arriving at any intersection), it turns <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">left</em> the first time, goes <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">straight</em> the second time, turns <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">right</em> the third time, and then repeats those directions starting again with <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">left</em> the fourth time, <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">straight</em> the fifth time, and so on.
```python
from pprint import pprint as pprint
import numpy as np
```

```python
tracks = []
carts = {}
with open('input.txt', 'r') as f:
    for y, line in enumerate(f):
        row = []
        for x, c in enumerate(line):
            if c in ' \/|-+':
                row.append(c)
            else:
                if c in '^v':
                    row.append('|')
                    carts[tuple([x,y])] = [c, 0]
                elif c in '<>':
                    row.append('-')
                    carts[tuple([x,y])] = [c, 0]
        tracks.append(row)
```

```python
print(len(tracks), len(tracks[0]))
for c in tracks[:20]:
    print(c[:20])
pprint(carts)
```

```python
direction_map = {'>': [1,0], 'v': [0,1], '<': [-1,0], '^': [0,-1]}
arrows = ['>', 'v', '<', '^']
collision = False
cur_carts = carts.copy()
while not collision:
    new_carts = cur_carts.copy()
    for cart in cur_carts:
        arrow = cur_carts[cart][0]
        turn_count = cur_carts[cart][1]
        new_pos = [sum(x) for x in zip(cart, direction_map[arrow])]
        if (np.array(new_pos) == np.array(list(new_carts.keys()))).all(1).any():
            print('EXPLOSION')
            print(new_pos)
            collision = True
        else:
            new_track = tracks[new_pos[1]][new_pos[0]]
            if new_track == '/':
                arrow = arrows[3 - arrows.index(arrow)]
            elif new_track == '\\':
                arrow = arrows[(5 - arrows.index(arrow))%4]
            elif new_track == '+':
                arrow = arrows[(arrows.index(arrow) + (turn_count-1)) % len(arrows)]
                turn_count = (turn_count+1) % 3
            new_carts.pop(cart)
            new_carts[tuple(new_pos)] = [arrow, turn_count]
    cur_carts = new_carts
```

```python
direction_map = {'>': [1,0], 'v': [0,1], '<': [-1,0], '^': [0,-1]}
arrows = ['>', 'v', '<', '^']
cur_carts = carts.copy()
tracker = 0
while len(cur_carts) > 1:
    new_carts = cur_carts.copy()
    for cart in sorted(cur_carts):
        if cart not in new_carts:
            break
        arrow = cur_carts[cart][0]
        turn_count = cur_carts[cart][1]
        new_pos = [sum(x) for x in zip(cart, direction_map[arrow])]
        if (np.array(new_pos) == np.array(list(new_carts.keys()))).all(1).any():
            print('EXPLOSION at T+', tracker, sep='')
            print(len(new_carts))
            pprint(new_carts)
#             print(len(new_carts))
#             pprint(new_carts)
            print("popping:", tuple(new_pos), new_carts.pop(tuple(new_pos)), 'and', cart, new_carts.pop(cart))
#             print(len(cur_carts))
#             pprint(cur_carts)
            print(len(new_carts))
            pprint(new_carts)
            
        else:
            new_track = tracks[new_pos[1]][new_pos[0]]
            if new_track == '/':
                arrow = arrows[3 - arrows.index(arrow)]
            elif new_track == '\\':
                arrow = arrows[(5 - arrows.index(arrow))%4]
            elif new_track == '+':
                arrow = arrows[(arrows.index(arrow) + (turn_count-1)) % len(arrows)]
                turn_count = (turn_count+1) % 3
            new_carts.pop(cart)
            new_carts[tuple(new_pos)] = [arrow, turn_count]
    tracker += 1
    cur_carts = new_carts
print(cur_carts)
```
