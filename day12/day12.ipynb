<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Your puzzle input also contains some notes you find on a nearby table: someone has been trying to figure out how these plants <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">spread</em> to nearby pots.  Based on the notes, for each generation of plants, a given pot has or does not have a plant based on whether that pot (and the two pots on either side of it) had a plant in the last generation. These are written as <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">LLCRR =&gt; N</code>, where <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">L</code> are pots to the left, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">C</code> is the current pot being considered, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">R</code> are the pots to the right, and <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">N</code> is whether the current pot will have a plant in the next generation. For example:</p>
<ul style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">A note like <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">..#.. =&gt; .</code> means that a pot that contains a plant but with no plants within two pots of it will not have a plant in it during the next generation.</li>
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">A note like <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">##.## =&gt; .</code> means that an empty pot with two plants on each side of it will remain empty in the next generation.</li>
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">A note like <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">.##.# =&gt; #</code> means that a pot has a plant in a given generation if, in the previous generation, there were plants in that pot, the one immediately to the left, and the one two pots to the right, but not in the ones immediately to the right and two to the left.</li>
</ul>

```python
from pprint import pprint as pprint
```

```python
init_state = ''
rules = {}
with open('input.txt', 'r') as f:
    for i, line in enumerate(f):
        if i == 0:
            init_state = line.split()[2]
        elif line.strip() != '':
            rule, result = line.strip().split(' => ')
            rules[rule] = result
```

```python
pprint(rules)
```

```python
generations = 20
state = '.'*4 + init_state + '.'*20
for gen in range(generations):
    print(state)
    new_state = ''
    for i in range(2, len(state)-2):
        new_state += rules[state[i-2:i+3]]
    state = '..' + new_state + '..'
total = 0
for i, p in enumerate(state):
    if p == '#':
        total += i-4
print(total)
```
