<div style="background-color: rgb(15, 15, 35); border: 1px solid rgb(255, 255, 255); color: rgb(204, 204, 204); font-family: monospace; margin-left: -5px; margin-right: -20px; padding: 1rem;"><h2 style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255);">--- Day 15: Beverage Bandits ---</h2><p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Having perfected their hot chocolate, the Elves have a new problem: the <a href="https://en.wikipedia.org/wiki/Goblin" style="border-color: rgb(0, 153, 0); color: rgb(0, 153, 0);">Goblins</a> that live in these caves will do anything to steal it. Looks like they're here for a fight.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">You scan the area, generating a map of the walls (<code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">#</code>), open cavern (<code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">.</code>), and starting position of every Goblin (<code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">G</code>) and Elf (<code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">E</code>) (your puzzle input).</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Combat proceeds in <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">rounds</em>; in each round, each unit that is still alive takes a <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">turn</em>, resolving all of its actions before the next unit's turn begins. On each unit's turn, it tries to <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">move</em> into range of an enemy (if it isn't already) and then <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">attack</em> (if it is in range).</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">All units are very disciplined and always follow very strict combat rules. Units never move or attack diagonally, as doing so would be dishonorable. When multiple choices are equally valid, ties are broken in <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">reading order</em>: top-to-bottom, then left-to-right.  For instance, the order in which units take their turns within a round is the <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">reading order of their starting positions</em> in that round, regardless of the type of unit or whether other units have moved after the round started.  For example:</p>
<pre style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background-color: rgb(16, 16, 26); display: inline-block;"><code style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204); background-color: rgb(16, 16, 26);">                 would take their

<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Each unit begins its turn by identifying all possible <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">targets</em> (enemy units). If no targets remain, combat ends.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Then, the unit identifies all of the open squares (<code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">.</code>) that are <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">in range</em> of each target; these are the squares which are <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">adjacent</em> (immediately up, down, left, or right) to any target and which aren't already occupied by a wall or another unit. Alternatively, the unit might <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">already</em> be in range of a target. If the unit is not already in range of a target, and there are no open squares which are in range of a target, the unit ends its turn.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">If the unit is already in range of a target, it does not <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">move</em>, but continues its turn with an <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">attack</em>. Otherwise, since it is not in range of a target, it <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">moves</em>.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">To <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">move</em>, the unit first considers the squares that are <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">in range</em> and determines <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">which of those squares it could reach in the fewest steps</em>. A <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">step</em> is a single movement to any <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">adjacent</em> (immediately up, down, left, or right) open (<code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">.</code>) square. Units cannot move into walls or other units. The unit does this while considering the <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">current positions of units</em> and does <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">not</em> do any prediction about where units will be later. If the unit cannot reach (find an open path to) any of the squares that are in range, it ends its turn. If multiple squares are in range and <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">tied</em> for being reachable in the fewest steps, the square which is first in <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">reading order</em> is chosen. For example:</p>

<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Once the Goblins and Elf reach the positions above, they all are either in range of a target or cannot find any square in range of a target, and so none of the units can move until a unit dies.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">After moving (or if the unit began its turn in range of a target), the unit <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">attacks</em>.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">To <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">attack</em>, the unit first determines <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">all</em> of the targets that are <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">in range</em> of it by being immediately <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">adjacent</em> to it. If there are no such targets, the unit ends its turn. Otherwise, the adjacent target with the <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">fewest hit points</em> is selected; in a tie, the adjacent target with the fewest hit points which is first in reading order is selected.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">The unit deals damage equal to its <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">attack power</em> to the selected target, reducing its hit points by that amount. If this reduces its hit points to <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">0</code> or fewer, the selected target <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">dies</em>: its square becomes <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">.</code> and it takes no further turns.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Each <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">unit</em>, either Goblin or Elf, has <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">3</code> <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">attack power</em> and starts with <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">200</code> <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">hit points</em>.</p>

<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">After attacking, the unit's turn ends.  Regardless of how the unit's turn ends, the next unit in the round takes its turn.  If all units have taken turns in this round, the round ends, and a new round begins.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">The Elves look quite outnumbered. (Combat only ends when a unit finds no targets during its turn.)</p>

```python
puzzle = []
with open('input.txt', 'r') as f:
    for line in f:
        if line != '\n':
            puzzle.append(line.strip())

caves = np.zeros((len(puzzle),len(puzzle[0])))
init_gobbs = []
init_elves = []

for i, line in enumerate(puzzle):
    for j, c in enumerate(line):
        if c == '#':
            caves[i,j] = 1
        else:
            if c == 'G':
                init_gobbs.append(tuple((i,j)))
            elif c == 'E':
                init_elves.append(tuple((i,j)))
```
